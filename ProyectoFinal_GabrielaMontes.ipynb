```python
# Importe de Librerias de procesado y modelado #
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import mean_squared_error as mse
from sklearn.pipeline import Pipeline
from sklearn.preprocessing  import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn import datasets, linear_model

from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Se extrae las variable numericas #
train_data = train_x.select_dtypes(include=['number'])
train_data.head()
```

```python
test_data = test_x.select_dtypes(include=['number'])
test_data.head()
```

```python
# Se elimina la columna que falta en el dataset de entrenamiento #
train_data.dropna(axis=0, subset=['SalePrice'], inplace=True)
train_data1 = train_data['SalePrice']
train_data.drop(['SalePrice'], axis=1, inplace=True)
```

<a id="1"></a>
<h1 # <div style="padding:5px ; color:white ; margin:0 ; font-size:20px ; font-family:'Calibri', serif ; text-align:left ; display:fill ; border-radius:5px;border:#bbb2a1 solid;overflow:hidden">

📌Se debe eliminar la columna faltante en el dataset de testeo porque al momento de realizar la predición, genera error mencionando específicamente que la característica "SalePrice" está ausente en los datos de evaluación o predicción, pero estaba presente durante el entrenamiento.

```python
modelo = Pipeline([('scaler', StandardScaler()),('features', PolynomialFeatures()),
                    ('regression', LinearRegression(fit_intercept=True))])
```

```python
# Se realiza la busqueda de hiperparametros #
hiperparametros = {'features__degree': [1, 2, 3, 4]}
grilla = GridSearchCV(modelo, hiperparametros, refit=True, cv=5, scoring='neg_root_mean_squared_error')
grilla.fit(train_data, train_data1)
```

```python
# Se imprime el mejor polinomio #
print(grilla.best_params_)
```

```python
# Se verifica los resulatdos del Cross-Validation (CV) #
grilla.cv_results_
```

```python
# Se visualiza los resultados del CV de forma gráfica #
for ni, i in enumerate(grilla.cv_results_['params']):
  print(i['features__degree'],-grilla.cv_results_['mean_test_score'][ni])
plt.errorbar(grilla.cv_results_['param_features__degree'].data,(-grilla.cv_results_['mean_test_score']),yerr=grilla.cv_results_['std_test_score'])
plt.yscale('log')
plt.xlabel('grado')
plt.ylabel('RMSE en validacion')
```

📌 En el gráfico se puede observar que el grado 1, es el parámetro en donde el polinomio se hace mas pequeño. Igualmente, se refleja que si se cambia a otros polinomios de grados mayores, comienza una tendencia de error.  

```python
# Se entrena el modelo con el mejor modelo estimador #
modelo_lineal=grilla.best_estimator_
modelo_lineal.fit(train_data, train_data)
```

```python
# Se hace las predicciones con test_data, train_data
y_pred_test = modelo_lineal.predict(test_data)
y_pred_train = modelo_lineal.predict(train_data)
```

```python
print('Precisión del modelo usando la raíz del error cuadratico medio (RMSE):')

score_train_modelo_lineal = mean_squared_error(train_data,y_pred_train,squared=False)
score_test_modelo_lineal = mean_squared_error(test_data,y_pred_test,squared=False)

print ("Train:", score_train_modelo_lineal)
print ("Testeo:", score_test_modelo_lineal)
```

📌 En el calculo del conjunto de prueba para "train" y "test", el valor es extremadamente pequeño, lo que sugiere que el modelo se ajusta prácticamente perfectamente a los datos.

📌 Los valores de RMSE tan cercanos a cero indican que el modelo tiene un rendimiento excepcional tanto en el conjunto de entrenamiento como en el conjunto de prueba. Se puede sugerir que el modelo es capaz de hacer predicciones muy precisas y que se adapta bien tanto a los datos de entrenamiento como a nuevos datos.

```python
# Se calcula el valor absoluto medio MAE #
from sklearn.metrics import mean_absolute_error
print("MAE",mean_absolute_error(test_data,y_pred_test))
print("MAE",mean_absolute_error(train_data,y_pred_train))
```

```python
# Se calcula el valor cuadratico medio MSE #
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(test_data,y_pred_test))
print("MSE",mean_squared_error(train_data,y_pred_train))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(test_data,y_pred_test)
r3 = r2_score(train_data,y_pred_train)
print(r2)
print(r3)
```

```python
# Se crea el modelo de árbol de decisión #
modelo = Pipeline([
    ('regression', DecisionTreeRegressor())
])
# Se define los hiperparámetros para la búsqueda aleatoria #
hiperparametros = {
    'regression__max_depth': [3, 5, 10, 12, 15],
    'regression__min_samples_split': [2, 5, 10 , 15, 20],
    'regression__min_samples_leaf': [1, 2, 4, 6, 8, 10]
}
# Se realiza la búsqueda aleatoria con validación cruzada #
grilla = RandomizedSearchCV(modelo, hiperparametros, refit=True, cv=5,
                            scoring='neg_root_mean_squared_error', verbose=3)
grilla.fit(train_data, train_data)
```

**Factor de inflación de la varianza (Variance inflation factor)**

```python
# Se importan las librerias #
from statsmodels.regression.linear_model import OLS
from statsmodels.tools.tools import add_constant
```

```python
def variance_inflation_factors(exog_df):
    '''
    Parameters
    ----------
    exog_df : dataframe, (nobs, k_vars)
        design matrix with all explanatory variables, as for example used in
        regression.
    Returns
    -------
    vif : Series
        variance inflation factors
    '''
    exog_df = add_constant(exog_df)
    vifs = pd.Series(
        [1 / (1. - OLS(exog_df[col].values,
                       exog_df.loc[:, exog_df.columns != col].values).fit().rsquared)
         for col in exog_df],
        index=exog_df.columns,
        name='VIF'
    )
    return vifs
variance_inflation_factors(train_data)
```

Se evidencia que hay tres variables con un VIF muy elevado: *1stFlrSF* , *2ndFlrSF* , *GrLivArea*.

Esto significa que colinealidad es muy fuerte. Indica que las variables estan altamente correlacionada con otras variables y puede afectar significativamente las estimaciones de los coeficientes de regresión.
